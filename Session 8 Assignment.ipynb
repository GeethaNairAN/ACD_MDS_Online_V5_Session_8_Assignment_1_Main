```python
# Count distance to previous Zero. 
# Count the difference of the distance from the previous Zero or the start of the Series whichever is closer. 
import pandas as pd
import numpy as np

# Define a function to count to the previous zero.

def count_to_zero(df):
    list1=[]
    valuex=0
    for x,y in df.iterrows():
        if y['X'] == 0:
            valuex=x + 1        # If Value is zero, save position+1 to find the relative position of the next value. 
            list1.append(0)
        else:
            newval=[(x + 1 - valuex)] # If the value is not zero, find the difference from the previous zero. 
            list1+=newval
    df['Y']=list1
    return df

# Test with the dataframe in the question.
df1=pd.DataFrame({'X':[7,2,0,1,2,3,4,0,3,4]})
print('Result Dataframe is below:')
print(count_to_zero(df1))
```

```python
# Test the count to zero function with more dataframes
df2=pd.DataFrame({'X':[0,4,5,0,9,0,7,8,10,0]})
print(count_to_zero(df2))

df3=pd.DataFrame({'X':[4,4,5,0,0,0,7,8,10,0]})
print(count_to_zero(df3))

df4=pd.DataFrame({'X':[4,4,5,1,2,3,7,8,10,0]})
print(count_to_zero(df4))
```

```python
# Create a DatetimeIndex that contains each business day of 2015 and use it to index a Series of random numbers
# Create a DatetimeIndex with the Business days of 2015
import datetime as dt
dtindx=pd.DatetimeIndex(start='2015-01-01',end='2015-12-31', freq='B')
print('DatatimeIndex is: ')
print(dtindx)

# Index a series of random numbers using the above datetimeindex.
s=pd.Series(np.random.randn(len(dtindx)), index=dtindx)
print('Series of Random Numbers:')
print(s)
```

```python
# Find the sum of the values in s for every Wednesday
# Form a series with weekdays and the random values.
sdays=pd.Series(s.values,index=dtindx.day_name())
print('Series of weekdays and random numbers: ')
print(sdays)

# Get the sum of random numbers for all Wednesdays
print('Sum of random numbers for all Wednesdays in the series: ', sdays.loc['Wednesday'].sum())
```

```python
# Average For each calendar month
print('Average for each calendar month:')
print(s.groupby(dtindx.month).mean())
```

```python
# For each group of four consecutive calendar months in s, find the date on which the highest value occurred.
# Find the maximum value for a month and form a series. 

scal_mon=pd.Series(s.groupby(dtindx.month).max().values,index=s.groupby(dtindx.month).idxmax())
print(scal_mon)

# Find the highest for every 4 consecutive months. 
# Form windows with 4 consecutive months and find the maximum from every window. 
print('Dates for which the highest value occurred in 4 consecutive months:')
for i in range(9):
    highval=scal_mon[i:4+i]
    print(highval.idxmax().date())
```
